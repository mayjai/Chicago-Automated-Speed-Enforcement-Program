# Violations & Crashes Data Cleaning

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import pandas as pd
import numpy as np
```

Load the data

```python
df_vio = pd.read_csv('Speed_Camera_Violations.csv')
df_crash = pd.read_csv('Traffic_Crashes_-_Crashes.csv')
```

This is how the origial <b>Violations Data</b> looks like

```python
df_vio
```

Delete unnecessary columns

```python
df_vio = df_vio.drop(columns=['X COORDINATE','Y COORDINATE','LOCATION','Historical Wards 2003-2015',\
                      'Zip Codes','Community Areas','Census Tracts'])
```

```python
df_vio
```

This is how the origial <b>Crashes Data</b> looks like, with a lot of unnecessary columns.

```python
df_crash
```

```python
df_crash.columns
```

Delete unnecessary columns

```python
df_crash = df_crash.drop(columns=['RD_NO','CRASH_DATE_EST_I','TRAFFIC_CONTROL_DEVICE','DEVICE_CONDITION',\
                      'LIGHTING_CONDITION','FIRST_CRASH_TYPE','TRAFFICWAY_TYPE','LANE_CNT','ALIGNMENT', \
                      'ROADWAY_SURFACE_COND', 'ROAD_DEFECT', 'REPORT_TYPE','INTERSECTION_RELATED_I',\
                      'NOT_RIGHT_OF_WAY_I', 'HIT_AND_RUN_I', 'DATE_POLICE_NOTIFIED','STREET_NO', 'STREET_DIRECTION',\
                      'STREET_NAME', 'BEAT_OF_OCCURRENCE','PHOTOS_TAKEN_I', 'STATEMENTS_TAKEN_I', 'DOORING_I', \
                      'WORK_ZONE_I','WORK_ZONE_TYPE', 'WORKERS_PRESENT_I', 'NUM_UNITS','MOST_SEVERE_INJURY',\
                      'INJURIES_TOTAL', 'INJURIES_FATAL','INJURIES_INCAPACITATING', 'INJURIES_NON_INCAPACITATING',\
                      'INJURIES_REPORTED_NOT_EVIDENT', 'INJURIES_NO_INDICATION','INJURIES_UNKNOWN', 'CRASH_HOUR',\
                      'CRASH_DAY_OF_WEEK', 'CRASH_MONTH','LOCATION'])
```

```python
df_crash
```

Change <b>date</b> to datatime

```python
df_vio['VIOLATION DATE'] = pd.to_datetime(df_vio['VIOLATION DATE']).dt.date
```

```python
df_crash['CRASH_DATE'] = pd.to_datetime(df_crash['CRASH_DATE']).dt.date
```

Subset <b>speed-related</b> crashes data

```python
# Checking how many speed-related crashes data are in the dataset
df_crash.PRIM_CONTRIBUTORY_CAUSE.str.contains('SPEED').sum()
```

```python
df_crash_speed = df_crash.loc[df_crash.PRIM_CONTRIBUTORY_CAUSE.str.contains('SPEED') | \
                              df_crash.SEC_CONTRIBUTORY_CAUSE.str.contains('SPEED')]
```

This is how the <b>speed-related crashes</b> dataset looks like

```python
df_crash_speed
```

Merge the two datasets by <b>date</b> and <b>location</b> (latitude and longitude), using a left join

```python
df_vio = df_vio.rename(columns={'VIOLATION DATE' : 'DATE'})
```

```python
df_crash_speed = df_crash_speed.rename(columns={'CRASH_DATE' : 'DATE'})
```

```python
df_merge = pd.merge(df_vio, df_crash_speed, how ='left', on = ['DATE', 'LATITUDE', 'LONGITUDE'])
```

This is how the <b>merged</b> dataset looks like, with a lot of NA

```python
df_merge
```

Save them as csv files for further analysis

```python
df_merge.to_csv('Violations_Crashes.csv', index = False)
```

```python
df_vio.to_csv('Violations.csv', index = False)
```

```python
df_crash_speed.to_csv('Crashes_Speed.csv', index = False)
```
